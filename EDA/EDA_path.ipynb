```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
graph = pd.read_csv('..\path\data\graph.csv', sep = ';', names = ['start', 'end'])
graph.head(5)
```

```python
train = pd.read_csv('..\path\data\\train-weekends.csv', sep = ',', names = graph.index)
train.head(5)
```

```python
cov = np.cov(train.T)
eigvals, eigvecs = np.linalg.eig(cov)

with plt.style.context('nicePlots'):
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (20, 10))
    mat = ax1.matshow(cov)
    fig.colorbar(mat, ax = ax1, fraction = 0.046, pad = 0.04)
    ax1.set_title('Covariance matrix')

    eigplot = ax2.scatter(eigvals.real, eigvals.imag, c = 'purple')
    ax2.set_ylabel('Imaginary')
    ax2.set_xlabel('Real')
    ax2.set_title('Eigenvalues')
```
